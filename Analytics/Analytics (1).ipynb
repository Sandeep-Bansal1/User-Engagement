```python
# Convert the 'Access_Timestamp' column to datetime format for time series analysis
final_df['Access_Timestamp'] = pd.to_datetime(final_df['Access_Timestamp'], errors='coerce')

# Set 'Access_Timestamp' as the index
time_series_data = final_df.set_index('Access_Timestamp')

# Resample to get the number of visits per month
monthly_visits = time_series_data['Frequency'].resample('M').sum()

# Filter for 2023 and 2024
visits_2023 = monthly_visits['2023']
visits_2024 = monthly_visits['2024']

# Calculate YoY growth percentage
yoy_growth = ((visits_2024.values - visits_2023.values) / visits_2023.values) * 100

# Create a DataFrame for comparison
significant_differences = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    '2023 Visits': visits_2023.values,
    '2024 Visits': visits_2024.values,
    'YoY Growth (%)': yoy_growth
})

# Print the DataFrame
print(significant_differences)

# Create a bar chart to visualize YoY Growth %
plt.figure(figsize=(14, 7))
plt.bar(significant_differences['Month'], significant_differences['YoY Growth (%)'])
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.title('Year-over-Year (YoY) Growth in Monthly User Visits (2023 vs. 2024)')
plt.xlabel('Month')
plt.ylabel('YoY Growth (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Convert Access_Timestamp to datetime if it's not already
final_df['Access_Timestamp'] = pd.to_datetime(final_df['Access_Timestamp'], errors='coerce')

# Extract various seasonality features
final_df['Year'] = final_df['Access_Timestamp'].dt.year
final_df['Month'] = final_df['Access_Timestamp'].dt.month
final_df['DayOfWeek'] = final_df['Access_Timestamp'].dt.dayofweek  # 0 = Monday, 6 = Sunday
final_df['Quarter'] = final_df['Access_Timestamp'].dt.quarter

# Custom Season Feature
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

final_df['Season'] = final_df['Month'].apply(get_season)

final_df.head()
```

```python
# Group by 'season' and count the number of users
season_counts = final_df.groupby('Season')['Access_Timestamp'].count()
```

```python
import matplotlib.pyplot as plt

# Data for plotting
seasons = season_counts.index.tolist()
user_counts = season_counts.values.tolist()
colors = ['#66b3ff','#99ff99','#ffcc99','#ff9999']
explode = (0.05, 0.05, 0.05, 0.05)  # Slightly explode all slices for better visibility

# Create the donut chart
fig, ax = plt.subplots(figsize=(8, 8))
wedges, texts, autotexts = ax.pie(
    user_counts,
    labels=seasons,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=explode,
    wedgeprops=dict(width=0.3, edgecolor='w')
)

# Customize text sizes
for text in texts:
    text.set_fontsize(12)

for autotext in autotexts:
    autotext.set_fontsize(12)

# Add a title
plt.title('Number of Users Who Reached Bottom per Season', fontsize=16,loc = 'right')

# Add a legend
ax.legend(
    wedges,
    seasons,
    title="Seasons",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1)
)

# Ensure the pie chart is circular
ax.axis('equal')

# Save the figure
plt.savefig('donut_chart_custom.png', bbox_inches='tight')

# Display the chart
plt.show()
```
